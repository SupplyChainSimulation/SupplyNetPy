```python
import simpy
import numpy as np
import matplotlib.pyplot as plt

simtime = 365
simruns = 1000

class UniformDist:
    """Class to generate uniform random numbers within a specified range."""
    def __init__(self, min_val, max_val):
        """Initialize the uniform distribution with minimum and maximum values."""
        self.min_val = min_val
        self.max_val = max_val
    def sample(self):
        """Generate a random number from the uniform distribution."""
        return np.random.uniform(self.min_val, self.max_val)
    
class NormalDist:
    """Class to generate normal random numbers with specified mean and standard deviation."""
    def __init__(self, mean, std_dev):
        """Initialize the normal distribution with mean and standard deviation."""
        self.mean = mean
        self.std_dev = std_dev
    def sample(self):
        """Generate a random number from the normal distribution."""
        e = -1
        while(e<0):
            e = np.random.normal(self.mean, self.std_dev)
        return e

uniform_arrival = UniformDist(3, 7)
uniform_order_size = UniformDist(50, 110)
normal_quantity = NormalDist(80, 10)

available_inventory = [0]
inventory_costs = [0]
inventory_spends = [0]
tranportation_costs = [0]
total_costs = [0]
total_units_sold = [0]

mouse = scm.Product(ID="MW11", name="Toad M11", manufacturing_cost=150, 
                             manufacturing_time=0, sell_price=300, buy_price=150,
                             raw_materials=[], units_per_cycle=2500)

#demands_list[0]['order_arrival_model'] = uniform_arrival.sample
demands_list[0]['order_quantity_model'] = normal_quantity.sample

for run in range(0, simruns):
    scm.global_logger.disable_logging()
    supplychainnet = scm.create_sc_net(nodes_list, links_list, demands_list)
    supplychainnet = scm.simulate_sc_net(supplychainnet, sim_time=simtime)
    for node in supplychainnet['nodes']:
        if('supplier' in node.node_type):
            continue
        available_inventory[-1] += node.inventory.inventory.level
        inventory_costs[-1] += node.inventory_cost        
        inventory_spends[-1] += sum([x[1] for x in node.inventory.inventory_spend])        
        tranportation_costs[-1] += sum([x[1] for x in node.transportation_cost])        
        total_costs[-1] += node.node_cost

    for demand in supplychainnet['demand']:
        if(demand.transportation_cost):
            tranportation_costs[-1] += demand.node_cost            
            total_units_sold[-1] += demand.total_products_sold            
            total_costs[-1] += demand.node_cost
    
    available_inventory.append(0)
    inventory_costs.append(0)
    inventory_spends.append(0)
    tranportation_costs.append(0)
    total_costs.append(0)
    total_units_sold.append(0)
```

```python
print(f"Performance:")
print(f"\tAvailable Inventory:\tmu = {np.mean(available_inventory):.2f},\t\t std = {np.std(available_inventory):.2f},\t\t std_err = {np.std(available_inventory)/np.sqrt(len(available_inventory)):.2f}")
print(f"\tInventory cost: \t mu = {np.mean(inventory_costs):.2f},\t std = {np.std(inventory_costs):.2f},\t std_err = {np.std(inventory_costs)/np.sqrt(len(inventory_costs)):.2f}")
print(f"\tInventory Spend:\t mu = {np.mean(inventory_spends):.2f},\t std = {np.std(inventory_spends):.2f},\t std_err = {np.std(inventory_spends)/np.sqrt(len(inventory_spends)):.2f}")
print(f"\tTransportation cost:\t mu = {np.mean(tranportation_costs):.2f},\t\t std = {np.std(tranportation_costs):.2f},\t\t std_err = {np.std(tranportation_costs)/np.sqrt(len(tranportation_costs)):.2f}")

print(f"\tTotal cost:\t\t mu = {np.mean(total_costs):.2f},\t std = {np.std(total_costs):.2f},\t std_err = {np.std(total_costs)/np.sqrt(len(total_costs)):.2f}")
print(f"\tTotal #products sold:\t mu = {np.mean(total_units_sold):.2f},\t std = {np.std(total_units_sold):.2f},\t\t std_err = {np.std(total_units_sold)/np.sqrt(len(total_units_sold)):.2f}")
print(f"\tSell price: {mouse.sell_price:.2f}")
revenue = np.mean(total_units_sold) * mouse.sell_price
print(f"\tTotal revenue: {revenue}")
print(f"\tNet profit (Totalcost - Revenue): {revenue - np.mean(total_costs)}")
```
